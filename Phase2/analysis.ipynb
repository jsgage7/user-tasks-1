### 1.1 Gender Distribution of Winners

```python
# Gender distribution
survivor_gender_counts = survivor_df['Gender'].value_counts()
idol_gender_counts = idol_df['Gender'].value_counts()

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot Survivor gender distribution
survivor_gender_counts.plot.pie(autopct='%1.1f%%', ax=ax1, colors=['#ff9999','#66b3ff'])
ax1.set_title('Survivor Winners by Gender', fontsize=14)
ax1.set_ylabel('')

# Plot American Idol gender distribution
idol_gender_counts.plot.pie(autopct='%1.1f%%', ax=ax2, colors=['#ff9999','#66b3ff'])
ax2.set_title('American Idol Winners by Gender', fontsize=14)
ax2.set_ylabel('')

plt.tight_layout()
plt.show()

# Print gender statistics
print("Survivor winners gender distribution:")
print(survivor_gender_counts)
print(f"Percentage female: {survivor_gender_counts['Female']/len(survivor_df)*100:.1f}%")

print("\nAmerican Idol winners gender distribution:")
print(idol_gender_counts)
print(f"Percentage female: {idol_gender_counts['Female']/len(idol_df)*100:.1f}%")
```

### 1.2 Age Distribution of Winners

```python
# Age distribution
plt.figure(figsize=(12, 6))

# Create boxplots for age comparison
age_data = [survivor_df['Winner_Age'], idol_df['Winner_Age']]
plt.boxplot(age_data, labels=['Survivor', 'American Idol'])
plt.title('Age Distribution of Winners', fontsize=16)
plt.ylabel('Age at Time of Win', fontsize=12)
plt.grid(True, alpha=0.3)

# Add individual points for better visualization
for i, data in enumerate(age_data):
    # Add jitter to x position
    x = np.random.normal(i+1, 0.04, size=len(data))
    plt.scatter(x, data, alpha=0.6)

plt.tight_layout()
plt.show()

# Print age statistics
print("Survivor winners age statistics:")
print(f"Average age: {survivor_df['Winner_Age'].mean():.1f} years")
print(f"Youngest winner: {survivor_df['Winner_Age'].min()} years")
print(f"Oldest winner: {survivor_df['Winner_Age'].max()} years")

print("\nAmerican Idol winners age statistics:")
print(f"Average age: {idol_df['Winner_Age'].mean():.1f} years")
print(f"Youngest winner: {idol_df['Winner_Age'].min()} years")
print(f"Oldest winner: {idol_df['Winner_Age'].max()} years")
```

### 1.3 Background/Region of Winners

```python
# Background analysis
# Extract state/region from background
def extract_region(background):
    parts = background.split(', ')
    if len(parts) > 1:
        return parts[-1]
    return 'Unknown'

survivor_df['Region'] = survivor_df['Winner_Background'].apply(extract_region)
idol_df['Region'] = idol_df['Winner_Background'].apply(extract_region)

# Count regions
survivor_regions = survivor_df['Region'].value_counts().head(8)
idol_regions = idol_df['Region'].value_counts().head(8)

# Plot region distribution
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

survivor_regions.plot.bar(ax=ax1, color='blue')
ax1.set_title('Top Regions of Survivor Winners', fontsize=14)
ax1.set_ylabel('Number of Winners', fontsize=12)
ax1.tick_params(axis='x', rotation=45)

idol_regions.plot.bar(ax=ax2, color='red')
ax2.set_title('Top Regions of American Idol Winners', fontsize=14)
ax2.set_ylabel('Number of Winners', fontsize=12)
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Print region statistics
print("Top regions for Survivor winners:")
print(survivor_regions)

print("\nTop regions for American Idol winners:")
print(idol_regions)
```

## 2. Viewership Trends Over Time

```python
plt.figure(figsize=(12, 6))

# Plot Survivor viewership
plt.plot(survivor_df['Year'], survivor_df['Avg_Viewership_Millions'], 
         marker='o', linestyle='-', color='blue', label='Survivor')

# Plot American Idol viewership
plt.plot(idol_df['Year'], idol_df['Avg_Viewership_Millions'], 
         marker='s', linestyle='-', color='red', label='American Idol')

plt.title('Average Viewership Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Viewership (Millions)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

# Analysis of viewership trends
survivor_peak = survivor_df.loc[survivor_df['Avg_Viewership_Millions'].idxmax()]
idol_peak = idol_df.loc[idol_df['Avg_Viewership_Millions'].idxmax()]

print(f"Survivor peak viewership: {survivor_peak['Avg_Viewership_Millions']} million in {survivor_peak['Year']} (Season {survivor_peak['Season']})")
print(f"American Idol peak viewership: {idol_peak['Avg_Viewership_Millions']} million in {idol_peak['Year']} (Season {idol_peak['Season']})")

# Calculate decline from peak to most recent season
survivor_latest = survivor_df.iloc[-1]
idol_latest = idol_df.iloc[-1]

survivor_decline = ((survivor_peak['Avg_Viewership_Millions'] - survivor_latest['Avg_Viewership_Millions']) / 
                    survivor_peak['Avg_Viewership_Millions']) * 100
idol_decline = ((idol_peak['Avg_Viewership_Millions'] - idol_latest['Avg_Viewership_Millions']) / 
                idol_peak['Avg_Viewership_Millions']) * 100

print(f"\nSurvivor viewership decline from peak: {survivor_decline:.1f}%")
print(f"American Idol viewership decline from peak: {idol_decline:.1f}%")
```

## 3. Analysis of How Both Shows Have Evolved Over Time

```python
# Plot number of contestants over time
plt.figure(figsize=(12, 6))

plt.plot(survivor_df['Season'], survivor_df['Contestants'], 
         marker='o', linestyle='-', color='blue', label='Survivor')
plt.plot(idol_df['Season'], idol_df['Contestants'], 
         marker='s', linestyle='-', color='red', label='American Idol')

plt.title('Number of Contestants per Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Contestants', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)

# Set y-axis to use integer ticks since we're counting people
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))

plt.tight_layout()
plt.show()

# Analyze location changes for Survivor
location_counts = survivor_df['Location'].value_counts()
print("Most frequent Survivor locations:")
print(location_counts.head(5))

# Analyze judge changes for American Idol
# Extract unique judge combinations
unique_judge_combos = idol_df['Judges'].unique()
print(f"\nAmerican Idol had {len(unique_judge_combos)} different judge combinations over {idol_df.shape[0]} seasons")

# Note about unique winners
print("\nNote on unique winners:")
print(f"American Idol has had 21 unique winners in 21 seasons (100% unique), while Survivor has had 42 unique winners in 44 seasons (95.5%), with Sandra Diaz-Twine and Tony Vlachos each winning twice.")
```
